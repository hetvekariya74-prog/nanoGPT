```python
import torch
import torch.nn as nn
from torch.nn import functional as F
```

```python
# read to inspect the dataset
with open("input.txt", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
# unique characters that occur in the dataset
chars = sorted(list(set(text)))
vocab_size = len(chars)
print("".join(chars))
print(vocab_size)
```

```python
# creating mapping from characters to integers
# we are building character level model, we will work on individual characters

stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}

# tokenizer
encode = lambda s: [stoi[c] for c in s] # encoder: takes a string, output a list of integers
decode = lambda l: "".join([itos[i] for i in l]) # decoder: takes a list of integers, outputs a string

print(encode("hii there"))
print(decode(encode("hii there")))
```

```python
# let's now tokenize the entire text of dataset and store it into tesor
data = torch.tensor(encode(text), dtype=torch.long)
print(data.shape, data.dtype)
print(data[:1000]) # the data that we look earlier will look like this into GPT
```

```python
# let's now split up the data into train and val sets
n = int(0.9*len(data)) # first 90% will be train, rest val
train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8 # aka context length
data[:block_size + 1]
```

```python
x = train_data[:block_size]
y = train_data[1:block_size+1]
for t in range(block_size):
    context = x[:t+1]
    target = y[t]
    print(f"when input is {context} the target is : {target}")
```

```python
torch.manual_seed(1337)
batch_size = 4 # how many independet sequences we will process in parallel
block_size = 8 # what is the maximum context length of the predictions?

def get_batch(split):
    # generate a small batch of data of inputs x and targets y
    data = train_data if split == "train" else val_data
    ix = torch.randint(len(data) - block_size, (batch_size, ))
    # print("ix")
    # print(ix)
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1: i+block_size+1] for i in ix])
    return x, y

xb, yb = get_batch("train")
print("inputs:")
print(xb.shape)
print(xb)
print("targets: ")
print(yb.shape)
print(yb)

print("------")

for b in range(batch_size): # batch dimension
    for t in range(block_size): # time dimention
        context = xb[b, :t+1]
        target = yb[b,t]
        print(f"when input is {context} the target is : {target}")
```

this contains 32 examples packed into a single batch. 

```python
# our input to transformer
print(xb)
```

```python
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        # each token directly reads off the logits for the next token from a lookup table
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)
        
    def forward(self, idx, targets=None):
        # idx = inputs
        # idx and targets are both (B, T) tensors of integers
        logits = self.token_embedding_table(idx) # B,T,C
        # we will interpret the value of lookup table as logits which will corresponds to the score for the next character in the sequence
        if targets is None:
            loss=None
        else:
            B,T,C = logits.shape
            targets = targets.view(B*T)
            loss = F.cross_entropy(logits.view(B*T,C), targets) # need to reshape logits, because pytorch expects `C` to be 2nd dimension 
        return logits, loss
    
    def generate(self, idx, max_new_tokens):
        # idx is (B,T) array of indices in the current context
        for _ in range(max_new_tokens):
            # get the predictions
            logits, loss = self(idx)
            # focus only on last time step
            logits = logits[:, -1, :] # becomes (B, C)
            # apply softmax to get probabilities
            probs = F.softmax(logits, dim=1) # B,C
            # sample from the distribution
            idx_next = torch.multinomial(probs, num_samples=1) # B,1
            # append sampled index to the running sequence
            idx = torch.cat((idx, idx_next), dim=1) # B, T+1
        return idx 
    
m = BigramLanguageModel(vocab_size)
logits, loss= m(xb, yb)
print(logits.shape)
print(loss)
```

How this works is: "24" will go the embedding table and will pluck out the 24th row. This row will be of dim `vocab_size`. Since we have 8 examples in one row, the output will be of dim `8 x vocab_size`. And we have 4 such examples in one batch, so the final output will be of dim `4 x 8 x vocab_size`.
Then 43 will go to the embedding table and will pluck out the 43rd row. This is how the embedding table works.

```python
idx = torch.zeros((1,1), dtype=torch.long)
print(decode(m.generate(idx, max_new_tokens=100)[0].tolist())) # since generation will be in batches, we need to pluck the first row and then decode to get back text
```

The mathematical trick in self-attention

```python
# consider following toy example:
torch.manual_seed(1337)
B, T, C = 4, 8, 2 # batch, time, channel
x = torch.randn(B,T,C)
x.shape
```

```python
# best way to communicate with the previous token is to take avg of previous tokens till the current token

# we want x[b,t] = mean_{i <= t} x x[b,i]
xbow = torch.zeros((B,T,C)) # bow = bag of words
for b in range(B):
    for t in range(T):
        xprev = x[b, :t+1] #(t,C)
        xbow[b, t] = torch.mean(xprev, 0) 
```

```python
# we wrote 2 for loops which are very inefficient, 
# much more efficient way is matrix multiplication, that's the trick

# consider toy example
torch.manual_seed(42)
a = torch.ones((3,3))
b = torch.randint(0, 10, (3,2)).float()
c = a @ b
print("a=")
print(a)
print("--")
print("b=")
print(b)
print("--")
print("c=")
print(c)
print("--")
```

```python
torch.manual_seed(42)
a = torch.tril(torch.ones((3,3))) # tril return lower triangular part
b = torch.randint(0, 10, (3,2)).float()
c = a @ b
print("a=")
print(a)
print("--")
print("b=")
print(b)
print("--")
print("c=")
print(c)
print("--")
```

see how `c` has been calculated in the cell above. `2`, then `2+6` then `2+6+6`. So like it keeps on adding the previous values. This is the trick in self-attention. 
Now similarly we can calculate average also, just by normalizing the values i.e dividing each row by the sum of the rows

```python
torch.manual_seed(42)
a = torch.tril(torch.ones((3,3))) # tril return lower triangular part
a = a / torch.sum(a, dim=1, keepdim=True)
b = torch.randint(0, 10, (3,2)).float()
c = a @ b
print("a=")
print(a)
print("--")
print("b=")
print(b)
print("--")
print("c=")
print(c)
print("--")
```

```python
# optimizing above for loop
wei = torch.tril(torch.ones(T,T))
wei = wei / wei.sum(1, keepdim=True) # wei = a
xbow2 = wei @ x # (T,T) @ (B, T, C) ----> (B, T, C)
# xbow2 will be identical to xbow
```

```python
xbow2[0]
```

```python
# another version of writing the same thing: using softmax
tril = torch.tril(torch.ones(T, T))
wei = torch.zeros((T,T))
wei = wei.masked_fill(tril == 0, float("-inf")) # in wei matrix, where tril == 0, make it -ve inf ; # future can't communicate with the past
wei = F.softmax(wei, dim=-1) # it is also normalization operation, so we get similarly wei matrix
xbow3 = wei @ x
```

```python
torch.allclose(xbow2, xbow3)
```

self attention

```python
# self attention
torch.manual_seed(1337)
B, T, C = 4, 8, 32 # batch, time, channel
x = torch.randn(B,T,C)

tril = torch.tril(torch.ones(T, T))
wei = torch.zeros((T,T))
wei = wei.masked_fill(tril == 0, float("-inf"))
wei = F.softmax(wei, dim=-1)
out = wei @ x

out.shape
```

Currently we are initialzing weight matrices with zeros, but we don't want it to be all uniform. We want it to be data dependent. We want to gather information from the past but want to do it in a data dependent way. This is the problem that self attention solves.

This is how self attention solves:
- Every single node/token at each position will emit two vectors: a query and a key vector.
  - Query vector: what am I looking for?
  - Key vector: what do I contain?
  - Then we take the dot product of the query and key vector. That dot product will now become our weights. 

```python
torch.manual_seed(1337)
B, T, C = 4, 8, 32 # batch, time, channel
x = torch.randn(B,T,C)

# implementing self attention for single head
head_size = 16 # hyperparameter
key = nn.Linear(C, head_size, bias=False)
query = nn.Linear(C, head_size, bias=False)
value = nn.Linear(C, head_size, bias=False)

k = key(x)   # B, T, 16 (head_size)
q = query(x) # B, T, 16

wei = q @ k.transpose(-2, -1) # B, T, 16 @ B, 16, T ----> B, T, T
tril = torch.tril(torch.ones(T, T))
wei = wei.masked_fill(tril == 0, float("-inf"))
wei = F.softmax(wei, dim=-1)

v = value(x)
out = wei @ v
# out = wei @ x

out.shape
```

NOTES:

- Attention is a communication mechanism.
- Attention simply acts over a set of vectors. There is no notion of space. This is why we need to positionally encode the tokens. 
- Each example across batch dimension is independent and never "talk" to each other.
- self-attention means that the keys and queries are produced from the same source. In cross-attention, the keys and queries are produced from different sources.
- "Scaled Attention" additionally divides wei by 1/sqrt(head_size). This makes it so when input Q,K are unit variance, wei will be unit variance too and Softmax will stay diffuse and not saturate too much. Illustration below.

```python
k = torch.randn(B,T, head_size)
q = torch.randn(B,T, head_size)
wei = q @ k.transpose(-2, -1)
```

Variance will be of head_size

```python
k = torch.randn(B,T, head_size)
q = torch.randn(B,T, head_size)
wei = q @ k.transpose(-2, -1) * (head_size)**-0.5
```

```python
k.var(), q.var(), wei.var()
```

If we divide by sqrt of head_size then var of wei will be 1.

```python
class LayerNorm1D:
    def __init__(self, dim, eps=1e-5):
        self.eps = eps
        self.gamma = torch.ones(dim)
        self.beta = torch.ones(dim)

    def __call__(self, x):
        # calculate the forward pass   
            # we normalize row-wise in layer norm
        xmean = x.mean(1, keepdim=True) # batch mean
        xvar = x.var(1, keepdim=True) # batch variance
        xhat = (x - xmean) / torch.sqrt(xvar + self.eps)
        self.out = self.gamma * xhat + self.beta
        
        return self.out
    
    def parameters(self):
        return [self.gamma, self.beta]
```
